```python
import pandas as pd

# Load the first CSV file into a DataFrame
file_01_path = '/content/01.csv'  # Update this path if necessary
df1 = pd.read_csv(file_01_path, encoding='latin1', dtype=str)

# Extract the relevant columns
df1_selected = df1[['Item : Internal ID', 'Item : Name', 'Subsidiary', 'Units']]

# Remove duplicate rows
df1_unique = df1_selected.drop_duplicates()

# Save the result to an Excel file
output_file_path = '/content/extracted_unique_items.csv'
df1_unique.to_csv(output_file_path, index=False)

# Display a message indicating where the file was saved
print(f"Extracted unique items saved to {output_file_path}")
```

```python
import pandas as pd

# Load the extracted unique items CSV file
unique_items_path = '/content/extracted_unique_items.csv'  # Update this path if necessary
df_unique_items = pd.read_csv(unique_items_path, encoding='latin1', dtype=str)

# Load the second CSV file into a DataFrame
file_02_path = '/content/02.csv'  # Update this path if necessary
df2 = pd.read_csv(file_02_path, encoding='latin1', dtype=str)

# Merge df2 with the unique items based on 'Item : Internal ID'
df2_updated = pd.merge(df2, df_unique_items, on='Item : Internal ID', how='left', suffixes=('', '_unique'))

# Update the 'Item : Name' and 'Units' columns in df2 with the corresponding values from the unique items DataFrame
df2_updated['Item : Name'] = df2_updated['Item : Name_unique'].combine_first(df2_updated['Item : Name'])
df2_updated['Units'] = df2_updated['Units_unique'].combine_first(df2_updated['Units'])
df2_updated['Subsidiary'] = df2_updated['Subsidiary_unique'].combine_first(df2_updated['Subsidiary'])

# Drop the temporary columns used for merging
df2_updated.drop(columns=['Item : Name_unique', 'Units_unique', 'Subsidiary_unique'], inplace=True)

# column_order = [
#     'Custom Form', 'Document Number', 'Customer Name', 'Date', 'Subsidiary',
#     'Class', 'Location', 'Department', 'Discount %', 'Discount', 'Amount Discount',
#     'Transaction Discount', 'Product Name', 'Item : Name', 'Quantity', 'Units',
#     'Tax Line', 'Item : Base Price', 'Amount (Transaction Total)', 'Amount (Tax)', 'Terms'
# ]
# #
# # Reorder the columns
# df2_updated = df2_updated[column_order]


# Save the updated DataFrame back to a CSV file
updated_file_path = '/content/updated_06.csv'
df2_updated.to_csv(updated_file_path, index=False)

# Display the first few rows of the updated DataFrame
# print("Updated DataFrame:")
# print(df2_updated.head())
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/LSTM_Cus_quantity (2).csv", parse_dates=True)
```

```python
df
```

```python
# Rename 'Unnamed: 0' column to 'Customer Name'
df.rename(columns={'Unnamed: 0': 'Customer Name'}, inplace=True)

# Drop 'Unnamed: 1' column
df.drop(columns=['Unnamed: 1'], inplace=True)

# Save the modified DataFrame back to a CSV file if needed
# df.to_csv('modified_file.csv', index=False)

# Display the first few rows of the modified DataFrame
print(df.head())
```

```python
# Display data type of 'Forecasted Value_ARIMA' column
print("Data Type of Forecasted Value_ARIMA Column:")
print(df['Forecasted Value_ARIMA'].dtype)

# Display one row of data from 'Forecasted Value_ARIMA' column
print("\nOne Row of Data from Forecasted Value_ARIMA Column:")
print(df['Forecasted Value_ARIMA'].head(1))
```

```python
# Convert 'Forecasted Value_ARIMA' column to string type
df['Forecasted Value_ARIMA'] = df['Forecasted Value_ARIMA'].astype(str)

# Filter the DataFrame to keep rows where 'Forecasted Value_ARIMA' column contains '1.0'
df_filtered = df[df['Forecasted Value_ARIMA'].str.contains('1.0')]

# Display the filtered DataFrame
print("Filtered DataFrame:")
print(df_filtered)

# Save the filtered DataFrame back to a CSV file if needed
# df_filtered.to_csv('filtered_file.csv', index=False)
```
